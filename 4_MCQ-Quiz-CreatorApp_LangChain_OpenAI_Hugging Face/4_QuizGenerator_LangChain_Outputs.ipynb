```python
import PyPDF2
```

```python
file_p = "/content/yolov7paper.pdf"

pdf_reader=PyPDF2.PdfReader(file_p)
text = ""
for page in pdf_reader.pages:
    text += page.extract_text()
    break
```

```python
text
```

```python
text_file = open("Output.txt", "w", encoding="utf-8")
text_file.write(text)
text_file.close()
```
